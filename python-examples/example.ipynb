```python
import f3d
import f3d_ipython  # adds magic to automatically display `f3d.image` objects
from turntable import download_file
```

```python
f3d.engine.autoload_plugins()

engine = f3d.engine(f3d.window.NATIVE_OFFSCREEN)
engine.window.size = 512, 400

model_fn = download_file("https://github.com/KhronosGroup/glTF-Sample-Models/raw/master/2.0/Buggy/glTF-Binary/Buggy.glb")
engine.loader.load_scene(model_fn)
engine.options.update(
    {
        "render.effect.tone-mapping": True,
        "render.effect.ambient-occlusion": True,
        "render.effect.translucency-support": True,
        "render.effect.anti-aliasing": True,
    }
)

engine.window.camera.position = 300, 300, 300

engine.window.render_to_image()
```

```python
engine.window.render_to_image(True)
```
